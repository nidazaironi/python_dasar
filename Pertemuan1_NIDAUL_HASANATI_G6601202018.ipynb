```python
import numpy as sp
traffic = sp.genfromtxt("/content/drive/MyDrive/Datasets/web_traffic.tsv",delimiter='\t')
print(traffic[:10])#  mencetak 10 baris pertama dari array traffic
print(traffic.shape)
```

```python
x = traffic[:,0] # 0: mengekstrak semua nilai di kolom pertama (kolom x) dari array traffic
y = traffic[:,1] # 1: mengekstrak semua nilai di kolom kedua (kolom y) dari array traffic
```

```python
x = x[~sp.isnan(y)] # menghapus elemen x yang memiliki pasangan y yang tidak valid (Not a Number / NaN).
y = y[~sp.isnan(y)] # menghapus nilai-nilai NaN di dalam array y itu sendiri.

# mengecek jumlah x dan y sama:
# print(x.shape)
# print(y.shape)

import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.show() # Menampilkan plot dan menghilangkan tulisan <matplotlib...> jika x adn y sama
```

```python
# Mempercantik tampilan grafik
plt.title("Web traffic last month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")
plt.xticks([w*7*24 for w in range(10)],['week %i' %w for w in range(10)])
plt.autoscale(tight=True)
plt.grid()
```

# latihan --intermezo

```python
y = np.array([2, 5, np.nan, 7, np.nan, 9])
print(y)
```

```python
sp.isnan(y)
# Output: array([False, False, True, False, True, False])
```

```python
sp.isnan(y) # negasi hasil awal
# Output: array([False, False, True, False, True, False])
```

```python
~sp.isnan(y)
# Output: array([False, False, True, False, True, False])
```
